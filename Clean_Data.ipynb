```python
data = data.drop('Unnamed: 0', axis = 'columns')
data = data.drop('Loan_ID', axis = 'columns')
data = data.drop('Total_Income', axis = 'columns')
```

```python
mode_val = data['Gender'].mode()[0]
data['Gender'] = data['Gender'].fillna(mode_val)

mode_val = data['Dependents'].mode()[0]
data['Dependents'] = data['Dependents'].fillna(mode_val)

mode_val = data['Self_Employed'].mode()[0]
data['Self_Employed'] = data['Self_Employed'].fillna(mode_val)

mode_val = data['Loan_Amount_Term'].mode()[0]
data['Loan_Amount_Term'] = data['Loan_Amount_Term'].fillna(mode_val)

mode_val = data['Credit_History'].mode()[0]
data['Credit_History'] = data['Credit_History'].fillna(mode_val)
```

```python
data.loc[data['Gender'] == 'Male', 'Gender'] = 1
data.loc[data['Gender'] == 'Female', 'Gender'] = 0

data.loc[data['Married'] == 'No', 'Married'] = 1
data.loc[data['Married'] == 'Yes', 'Married'] = 2

data.loc[data['Dependents'] == '0', 'Dependents'] = 0
data.loc[data['Dependents'] == '1', 'Dependents'] = 1
data.loc[data['Dependents'] == '2', 'Dependents'] = 1
data.loc[data['Dependents'] == '3+', 'Dependents'] = 1

data.loc[data['Education'] == 'Graduate', 'Education'] = 1
data.loc[data['Education'] == 'Not Graduate', 'Education'] = 0

data.loc[data['Self_Employed'] == 'No', 'Self_Employed'] = 1
data.loc[data['Self_Employed'] == 'Yes', 'Self_Employed'] = 0

data.loc[data['Loan_Status'] == 'Y', 'Loan_Status'] = 1
data.loc[data['Loan_Status'] == 'N', 'Loan_Status'] = 0

data.loc[data['Property_Area'] == 'Urban', 'Property_Area'] = 1
data.loc[data['Property_Area'] == 'Rural', 'Property_Area'] = 2
data.loc[data['Property_Area'] == 'Semiurban', 'Property_Area'] = 3

data[x]
```

```python
def splitData(dataframe, percent_train):
    m = data[x].shape[0]
    n_train = int(np.ceil(percent_train * m))
    n_test = m - n_train
    index_train = list(np.sort(np.random.choice(m, n_train, replace=False)))
    data_train = dataframe.iloc[index_train]

    ALL = set(range(m))
    TRAIN = set(index_train)
    TEST = ALL - TRAIN

    index_test = sorted(list(TEST))
    data_test = dataframe.iloc[index_test]
    
    return (data_train, data_test)


(data_train, data_test) = splitData(data[x], 0.7)
```
