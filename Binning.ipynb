```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb


dt = pd.read_pickle('Data_train.pickle')

dt
```

```python
fields = ['ApplicantIncome','CoapplicantIncome','LoanAmount','Loan_Amount_Term']
for f in fields:
    fbin = pd.qcut(dt[f], 10, duplicates = 'drop')
    print(fbin)
    fname = f.upper() + '_BIN'
    dt[fname] = fbin
    dt
    fig = plt.figure(figsize = (15,8))
    sb.countplot(data = dt, x = fname)
```
